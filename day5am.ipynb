```python
# query multiple kondisi dengan mengambil beberapa kolom saja

df[(df['year_id'] > 2005) &
   (df['fran_id'] == 'Wizards') &
   (df['pts'] >= 100)][['fran_id','year_id','pts']]
```

```python
# latihan query
# year_id [2000 - 2010]
# pts >=90
# game_result = W
```

```python
df[(df['year_id'] >= 2000) &
   (df['year_id'] <= 2010) &
   (df['pts'] >= 90) &
   (df['game_result'] == 'W')]
```

```python
df[(df['year_id'] >= 2000) &
   (df['year_id'] <= 2010) &
   (df['pts'] >= 90) &
   (df['game_result'] == 'W')].shape
```

```python
# query dengan fungsi 

df.query("year_id >= 2000 & year_id <= 2010 & pts >= 90 & game_result == 'W'")
```

```python
df.query("year_id >= 2000 & \
          year_id <= 2010 & \
          pts >= 90 & \
          game_result == 'W'").shape
```

```python
# query dengan fungsi dan kolom spesifik

df.query("year_id >= 2000 & \
          year_id <= 2010 & \
          pts >= 90 & \
          game_result == 'W'")[['team_id', 'fran_id']]
```
