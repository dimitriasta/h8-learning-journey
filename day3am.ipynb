```python
num = 40

if num == 30:
  print(f"nilai {num} sama dengan 30")
elif num > 30:
  print(f"nilai {num} lebih dari 30")
else:
  print(f"nilai {num} kurang dari 30")
```

```python
nilai = int(input("Berapa score yang didapatkan?"))

if nilai<40:
  print(f"dengan score {nilai} grade anda adalah E")
elif nilai>=40 and nilai<60:
  print(f"dengan score {nilai} grade anda adalah D")
elif nilai>=60 and nilai<70:
  print(f"dengan score {nilai} grade anda adalah C")
elif nilai>=70 and nilai<80:
  print(f"dengan score {nilai} grade anda adalah B")
else:
  print(f"dengan score {nilai} grade anda adalah A")
```

```python
nilai = int(input("Berapa score yang didapatkan?"))

if nilai>=0 and nilai<=100:
  if nilai<40:
    print(f"dengan score {nilai} grade anda adalah E")
  elif nilai>=40 and nilai<60:
    print(f"dengan score {nilai} grade anda adalah D")
  elif nilai>=60 and nilai<70:
    print(f"dengan score {nilai} grade anda adalah C")
  elif nilai>=70 and nilai<80:
    print(f"dengan score {nilai} grade anda adalah B")
  else:
    print(f"dengan score {nilai} grade anda adalah A")
else:
  print("score yang ada masukkan salah")
```

```python
i=1

while i<=10:
  print(f"5 x {i} = {5*i}")
  i=i+1
```

```python
#tabel perkalian lengkap

i=1
j=1

while i<=10:
  print(f'perkalian angka {i}')
  while j<=10:
    print(f'{i} * {j} = {i*j}')
    j=j+1
  print('_'*50)
  i=i+1
  j=1
```

```python
#continue

i=0

while i<=10:
  i=i+1
  if i in (7,8,9):
    continue
  print(i)
```

```python
#break

i=0

while i<=10:
  i=i+1
  print(i)
  if i == 7:
    break
```

```python
for i in range(1,11):
  print(f'5 x {i} = {5*i}')
```

```python
#table perkalian lengkap

for i in range(1,11):
  for j in range(1,11):
    print(f'{i} * {j} = {i*j}')
  print('-'*50)
```
